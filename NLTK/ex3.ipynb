```python
import nltk
import random
```

```python
nltk.download()
#Passo: d
#Passo: all
#Passo: q
```

```python
arquivo = open("/content/drive/MyDrive/Codigos/BootcampPythonPLN/nltk/corpus_teste.txt", 'r')
corpus = arquivo.read()
```

```python
tokens = nltk.word_tokenize(corpus)
```

#Letra A

```python
tokens_bigrams = list(nltk.bigrams(tokens))

print(tokens_bigrams)
```

#Letra B

```python
tokens_trigrams = list(nltk.trigrams(tokens))

print(tokens_trigrams)
```

#Letra C

```python
tokens_4grams = list(nltk.ngrams(tokens, 4))

print(tokens_4grams)
```

#Letra D

```python
entidadesNomeadasBigram = []
for bigrama in tokens_bigrams:
  if bigrama[0][0].isupper() and bigrama[1][0].isupper():
    entidadesNomeadasBigram.append(bigrama)

print(entidadesNomeadasBigram)

entidadesNomeadasTrigram = []
for trigrama in tokens_trigrams:
  if trigrama[0][0].isupper() and trigrama[1][0].isupper() and trigrama[2][0].isupper():
    entidadesNomeadasTrigram.append(trigrama)

print(entidadesNomeadasTrigram)
```

#Letra E

```python
stemmer = nltk.RSLPStemmer()

tokenizer = nltk.tokenize.RegexpTokenizer(r'[a-zA-Z]\w*')
palavras = tokenizer.tokenize(corpus)

qtdPalavras = len(palavras)

for i in range(3):
  rand = random.randint(0, qtdPalavras)
  print(f"Palavra: {palavras[rand]} -> Stemming: {stemmer.stem(palavras[rand])}")
```

#Letra F

```python
etiq_padrao = nltk.tag.DefaultTagger('N')
etiquetador = nltk.tag.UnigramTagger(sentencas_treino, backoff = etiq_padrao)
```

```python
sentencas = []
sentenca = []
for token in tokens:
  sentenca.append(token)
  if token == '.':
    sentencas.append(sentenca)
    sentenca = []

#tratamento especifico
for token in sentencas[0]:
  sentenca.append(token)
  if token == 'XLII' or token == 'temporada' or token == '09h49m':
    sentencas.insert(0, sentenca)
    sentenca = []

rand = random.randint(0, len(sentencas))
sentencas_treino = nltk.corpus.mac_morpho.tagged_sents()

etiquetado = etiquetador.tag(sentencas[rand])

print(f"Frase: {sentencas[rand]}")
print(etiquetado)
```

#Letra G

```python
etiquetadoFull = etiquetador.tag(tokens)

print(etiquetadoFull)
```

#Letra H

```python
#entidade nomeada seguindo o padrao <NPROP><NPROP> ou <N><N>
pattern = 'NP: {<NPROP><NPROP> | <N><N>}'
analise_gramatical = nltk.chunk.RegexpParser(pattern)

arvore = analise_gramatical.parse(etiquetadoFull)
print(arvore)
```
